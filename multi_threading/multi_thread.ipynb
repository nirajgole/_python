```python
#importing libraries

from time import sleep
from threading import *
```

```python
# defining object
class Hello(Thread):
    def run(self):
        for i in range(5):
            print('Hello ğŸ‘‹')
            sleep(1) # delaying process by one second

# defining object
class Hi(Thread):
    def run(self):
        for i in range(5):
            print('Hi ğŸ‘‹ğŸ»')
            sleep(1)  # delaying process by one second
```

```python
t1= Hello() #initializing object 
t2=Hi() #initializing object 
```

```python
t1.start() # run thread
sleep(0.2) #to avoid execution at same time
t2.start() # run thread
```

```python
# to start main thread after execution of t1 and t2 thread 
t1.join()
t2.join()
```
