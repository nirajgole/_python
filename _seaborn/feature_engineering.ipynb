label encoding

```python
from sklearn.preprocessing import LabelEncoder

e = df.copy()

# creating instance for LabelEncoder
le = LabelEncoder()

# loop over columns
for i in e.columns:
  # find all object columns as le will be performed on obj data
  if e[i].dtype=='object':
    e[i]=le.fit_transform(e[i])
e
```

one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

data = df.copy()


object_columns = data.select_dtypes(include='object').columns
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_columns = encoder.fit_transform(data[object_columns])

encoded_df = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out(object_columns))

data_encoded = data.drop(object_columns, axis=1)


data_encoded = pd.concat([data_encoded, encoded_df], axis=1)


data_encoded
```
