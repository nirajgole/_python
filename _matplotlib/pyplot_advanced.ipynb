```python
import matplotlib.pyplot as plt # for data visualization
import numpy as np # for array related operations
import pandas as pd # for working with CSV files
csv_data=pd.read_csv('../data/StandardMetropolitanAreasData.csv') # read data from CSV file
```

#### Plotting consecutive plots using Matplotlib

```python
plt.plot(csv_data.work_force,csv_data.income,color='red')
plt.plot(csv_data.physicians,csv_data.income,color='blue')
plt.show()
```

#### Making the Plot Descriptive

```python
plt.scatter(csv_data.percent_senior,csv_data.crime_rate,color='red')

#Adding a title to the plot
plt.title('Crime Rate vs. Percent Senior')
#adding a x-axis label
plt.xlabel('Percent Senior')
#adding a y-axis label
plt.ylabel('Crime Rate')
plt.show()
```

```python

plt.plot(csv_data.work_force,csv_data.income,color='red',label='Work Force')
plt.plot(csv_data.physicians,csv_data.income,color='blue',label='Physicians')
#Adding legends to the plot
plt.legend()
plt.show()
```

### Formatting the plot
#### Note. Read documentation for more info and formatting options

#### <ol><li>Changing Colors</li></ol>

```python
plt.plot(csv_data.physicians,csv_data.income,color='blue',label='Physicians') # specify color by name
plt.plot(csv_data.work_force,csv_data.income,color='c',label='Work Force') # short color code (rgbcmyk)
plt.plot(csv_data.hospital_beds,csv_data.income,color='0.75',label='Hospital Beds') # Grayscale between 0 and 1
plt.legend() # add legend to the plot
plt.show()

#? hex, RGB tuple and all HTML color codes are also supported
```

#### <ol start='2'><li>Changing Line Styles</li></ol>

```python
# specify line style by name
plt.plot(csv_data.work_force,csv_data.income,color='red',linestyle='dashed')
plt.xlabel('Work Force')
plt.ylabel('Income')
plt.show()
```

#### <ol start='3'><li>Adding Markers</li></ol>

```python
plt.plot(csv_data.work_force,csv_data.income,marker='o')
plt.xlabel('Work Force')
plt.ylabel('Income')
plt.show()
```

#### <ol start='4'><li>Shorthand</li></ol>


```python
# linestyle='--'
# marker='o'
# color='red'
#plt.plot(csv_data.work_force,csv_data.income,color='red',linestyle='--',marker='o')

# shorthand for the above is 'r--o'
plt.plot(csv_data.work_force,csv_data.income,'r--o')
plt.show()
```

### Changing Plot Size

```python
#plt.figure(figsize=(new_width_pixel,new_height_pixel))
plt.figure(figsize=(12,5)) # 12 x 5 plot
plt.plot(csv_data.work_force,csv_data.income)
plt.xlabel('Work Force')
plt.ylabel('Income')
plt.show()
```

#### Multiple Plots

```python
plt.subplot(1,2,1) # 1 row, 2 columns, 1st subplot
plt.plot(csv_data.work_force,csv_data.income,'go')
plt.title('Work Force vs. Income')

plt.subplot(1,2,2) # 1 row, 2 columns, 2nd subplot
plt.plot(csv_data.physicians,csv_data.income,'ro')
plt.title('Physicians vs. Income')

plt.show()
```

```python
# Using fig and axis to create subplots
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True) # 1 row, 2 columns, share y-axis
fig.suptitle('Horizontally stacked subplots')
ax1.plot(csv_data.work_force,csv_data.income,'go')
ax1.set_title('Work Force vs. Income')

ax2.plot(csv_data.physicians,csv_data.income,'ro')
ax2.set_title('Physicians vs. Income')
```

```python
# 2 rows 2 columns 2 x 2 grid
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(6,6))
ax[0,0].plot(csv_data.work_force,csv_data.income,'go')
ax[0,1].plot(csv_data.physicians,csv_data.income,'ro')
ax[1,0].plot(csv_data.hospital_beds,csv_data.income,'bo')
ax[1,1].plot(csv_data.percent_senior,csv_data.crime_rate,'yo')

plt.show()
```
