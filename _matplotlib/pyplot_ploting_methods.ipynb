### Scatter Graph
<ul>
<li>
The position of each dot on the horizontal and vertical axis indicates values for an individual data point.
</li>
<li>
Scatter plots are used to observe relationships between variables.
</li>
<li>A scatter plot can be useful for identifying other patterns in data.</li>
<li>We can divide data points into groups based on how closely sets of points cluster together.</li>
<li>Scatter plots can also show if there are any unexpected gaps in the data and if there are any outlier points. (Look at the 2 points away from rest of the data in the scatter plot. Those are outliers.)</li>
<li>This can be useful if we want to segment the data into different parts, like categorising users into different groups.</li>
</ul>

#### To create a scatter plot in Matplotlib we can use the .scatter() method:

```python
plt.scatter(csv_data.crime_rate,csv_data.percent_senior)
#! plt.show() is required to show the plot in IDE's like Spyder, PyCharm, VsCode, etc.
# will not required in Jupyter Notebook, IPython Notebook, Colab, etc.
plt.show() #showing the plot 
```

### Line Plot
<li>In Matplotlib we can create a line chart by calling the plot method.</li>
<li>plot() is a versatile command, and can take an arbitrary number of arguments.</li>
<li>If you provide a list of n elements to the .plot() function, matplotlib will assume it is a sequence of  y  values, and automatically generates the  x  values for you as a range of n elements starting from 0. </li>

```python
 # Simple line plot
 plt.plot([1,2,3,4]) # y-axis values, single argument
 plt.show()
```

```python
plt.plot(csv_data.work_force,csv_data.income)
# plt.plot(csv_data.work_force,csv_data.income,'.',color='red')
plt.show()
```

#### Histogram

<li>Histograms are a very common type of plots when we are looking at data like height and weight, stock prices, waiting time for a customer, etc which are continuous in nature.</li>
<li>Histograms are good for showing general distributional features of dataset variables. You can see roughly where the peaks of the distribution are, whether the distribution is skewed or symmetric, and if there are any outliers.</li>

```python
plt.hist(csv_data.crime_rate)
plt.show()
```

#### Pie Chart
A pie chart is best used when trying to work out the composition of something. If you have categorical data then using a pie chart would work really well as each slice can represent a different category. A good example of a pie chart can be seen below.

<img src='./pie_chart.png'/>

```python
categories=['Cat A','Cat B','Cat C','Cat D','Cat E']
share=[20,30,40,50,60]
# categories=csv_data.region.unique()
# share=csv_data.crime_rate.values

plt.pie(share,labels=categories,autopct='%1.1f%%',shadow=True,startangle=90,explode=[0,0,0,0,0.1],colors=['red','green','blue','yellow','orange'],textprops={'fontsize':15},counterclock=False,frame=True,radius=1.2,pctdistance=0.8,labeldistance=1.2,center=[0,0])
plt.show()
```
